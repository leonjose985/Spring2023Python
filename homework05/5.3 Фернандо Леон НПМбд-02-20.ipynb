## Exercise 1 
Полиномиальная регрессия

```python
import numpy as np
import numpy.linalg
import matplotlib.pyplot as plt

def f(x):
    return (np.sin(x/5) * np.exp(x/10)) + 5*np.exp(-1/2*x)

def approx(n):
    dots = np.linspace(1, 15, n)
    
    a = np.array([[x**j for j in range(n)] for x in dots])
    b = f(dots)
    
    w =np.linalg.solve(a, b)
    res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])
    return res 

x = np.linspace(1, 15, 250)
plt.grid()
plt.title('Исходная финкция')
plt.plot(x, f(x))
plt.xlabel('x')
plt.ylabel('y')
plt.show()

######################## n=5
n = 5
plt.grid()
plt.title(f'Точки, по которым строился многочлен при n = {n}')
plt.plot(np.linspace(1, 15, n), f(np.linspace(1, 15, n)), '8')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.grid()
plt.title(f'Полиномальная аппроксимация при n = {n}')
plt.plot(np.linspace(1, 15, n), approx(n))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
######################## n=15
n = 15
plt.grid()
plt.title(f'Точки, по которым строился многочлен при n = {n}')
plt.plot(np.linspace(1, 15, n), f(np.linspace(1, 15, n)), '8')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.grid()
plt.title(f'Полиномальная аппроксимация при n = {n}')
plt.plot(np.linspace(1, 15, n), approx(n))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
##################### n=100
n = 100
plt.grid()
plt.title(f'Точки, по которым строился многочлен при n = {n}')
plt.plot(np.linspace(1, 15, n), f(np.linspace(1, 15, n)), '8')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.grid()
plt.title(f'Полиномальная аппроксимация при n = {n}')
plt.plot(np.linspace(1, 15, n), approx(n))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Мы видим, что с увеличением числа n функция полиномиального приближения становится более похожей на исходную функцию

## Exercise 2
Полиномиальная регрессия с зашумлённым данными

```python
import numpy.random

def approx(n):
    dots = np.linspace(1, 15, n)
    
    a = np.array([[x**i for i in range(n)] for x in dots])
    b = f(dots) + np.random.normal(0, 0.01)
    
    w = np.linalg.solve(a, b)
    res = np.array([sum([w[i] * x**i for i in range(n)]) for x in dots])
    return res

x = np.linspace(1, 15, 250)
plt.grid()
plt.title('Исходная финкция')
plt.plot(x, f(x))
plt.xlabel('x')
plt.ylabel('y')
plt.show()

############################## n=5

n = 5
plt.grid()
plt.title(f'Точки, по которым строился многочлен при n = {n}')
plt.plot(np.linspace(1, 15, n), f(np.linspace(1, 15, n)), '8')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.grid()
plt.title(f'Полиномальная аппроксимация при n = {n}')
plt.plot(np.linspace(1, 15, n), approx(n))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
################################### n=15
n = 15
plt.grid()
plt.title(f'Точки, по которым строился многочлен при n = {n}')
plt.plot(np.linspace(1, 15, n), f(np.linspace(1, 15, n)), '8')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.grid()
plt.title(f'Полиномиальная аппроксимация с шумом при n = {n}')
plt.plot(np.linspace(1, 15, n), approx(n))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
################################### n=100
n = 100
plt.grid()
plt.title(f'Точки, по которым строился многочлен при n = {n}')
plt.plot(np.linspace(1, 15, n), f(np.linspace(1, 15, n)), '8')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.grid()
plt.title(f'Полиномиальная аппроксимация с шумом при n = {n}')
plt.plot(np.linspace(1, 15, n), approx(n))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Из графиков видно, что при введении шума точность полиномиального приближения ухудшается, и это больше влияет на приближения с малым n.

## Exercise 3
Метод наименьших квадратов 

```python
N = 100
n = 3
dots = np.linspace(1, 15, N)
a = np.array([[x**j for j in range(n)] for x in dots])
b = f(dots) + np.random.normal(0, 0.1)

w, resid, rank, s = numpy.linalg.lstsq(a, b)  # точки в полиноме
res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  # решение

x = np.linspace(1, 15, 250)
plt.grid()
plt.title('Исходная функция')
plt.plot(x, f(x))
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.grid()
plt.title(f'Точки, по которым строился многочлен при n = {n}')
plt.plot(dots, f(dots), '8')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.grid()
plt.title(f'Метод наименьших квадратов при n = {n}')
plt.plot(dots, res)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Как вы можете видеть, метод наименьших квадратов не дает точного графика при n=3. С увеличением n график становится более точным.
